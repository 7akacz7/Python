```python
import random

def zgadnij_liczbe():
    while True:
      
        szukana_liczba = random.randint(1, 100)
        liczba_prob = 0
        
        print("\n--- Nowa gra! ---")
        print("Wylosowałem liczbę od 1 do 100. Zgadnij jaka to liczba!")

        while True:
            try:
                wybor = int(input("Podaj swoją liczbę: "))
                liczba_prob += 1
                if wybor < szukana_liczba:
                    print("Za mało!")
                elif wybor > szukana_liczba:
                    print("Za dużo!")
                else:
                    print(f"Brawo! Odgadłeś liczbę {szukana_liczba} w {liczba_prob} próbach.")
                    break 
            except ValueError:
                print("To nie jest prawidłowa liczba. Spróbuj ponownie.")

        jeszcze_raz = input("\nCzy chcesz zagrać ponownie? (tak/nie): ").strip().lower()
        if jeszcze_raz not in ('tak', 't', 'yes', 'y'):
            print("Dzięki za grę! Do zobaczenia.")
            break 

zgadnij_liczbe()
```

```python
import string

def wyczysc_tekst(tekst):
    tabela_znakow = str.maketrans('', '', string.punctuation)
    return tekst.translate(tabela_znakow).lower()

def policz_slowa(tekst):
    slowa = tekst.split()
    return len(slowa)

def najdluzsze_slowo(tekst):
    czysty_tekst = wyczysc_tekst(tekst)
    slowa = czysty_tekst.split()
    
    if not slowa: 
        return ""
    
    return max(slowa, key=len)

def czestotliwosc_slow(tekst):
    czysty_tekst = wyczysc_tekst(tekst)
    slowa = czysty_tekst.split()
    
    slownik_wystapien = {}
    for slowo in slowa:
        slownik_wystapien[slowo] = slownik_wystapien.get(slowo, 0) + 1
        
    return slownik_wystapien

def cenzura(tekst, zakazane_slowa):
    slowa = tekst.split()
    wynik = []
    
   
    zakazane_slowa_lower = {slowo.lower() for slowo in zakazane_slowa}
    
    for slowo in slowa:
        sprawdzane_slowo = wyczysc_tekst(slowo)
        
        if sprawdzane_slowo in zakazane_slowa_lower:
            wynik.append("***")
        else:
            wynik.append(slowo)
            
    return " ".join(wynik)


# CZĘŚĆ TESTOWA

if __name__ == "__main__":
    testowy_tekst = """
    Językoznawca Maciej Grochowski wyliczył, że
    podstawowych wulgaryzmów w języku polskim jest tylko pięć.
    Są to "chuj", "pizda", "jebać", "pierdolić" i "kurwa".
    Reszta to kombinacje tych pięciu, wyrazy pochodne i związki frazeologiczne.
    Nowe wulgaryzmy pojawiają się, gdy zostaną 
    utworzone nowe konfiguracje wyrazów bądź też zmieni
    się ich kontekst semantyczny. Słownik polszczyzny rzeczywistej
    wyróżnia czterdzieści siedem funkcji wypełnianych przez wulgaryzmy
    """

    print("TEKST ORYGINALNY")
    print(testowy_tekst)

    print("\n TESTY FUNKCJI")
    
    liczba = policz_slowa(testowy_tekst)
    print(f"1. Liczba słów w tekście: {liczba}")

    najdluzsze = najdluzsze_slowo(testowy_tekst)
    print(f"2. Najdłuższe słowo: '{najdluzsze}' (długość: {len(najdluzsze)})")

    czestotliwosc = czestotliwosc_slow(testowy_tekst)
    print(f"3. Słowo 'python' występuje: {czestotliwosc.get('python', 0)} razy.")
    print(f"   Słowo 'programowania' występuje: {czestotliwosc.get('programowania', 0)} razy.")
    
    posortowane_slowa = sorted(czestotliwosc.items(), key=lambda x: x[1], reverse=True)[:5]
    print(f"   5 najczęstszych słów: {posortowane_slowa}")

    lista_zakazanych = ["chuj", "pierdolić", "kurwa", "jebać", "pizda"]
    ocenzurowany = cenzura(testowy_tekst, lista_zakazanych)
    
    print(f"\n4. Tekst po cenzurze słów {lista_zakazanych}:\n{ocenzurowany}")
```

```python
import csv

# 1. TWORZENIE PLIKU CSV

dane_studentow = [
    ["imię", "nazwisko", "ocena"], # Nagłówki
    ["Jan", "Kowalski", "4.5"],
    ["Anna", "Nowak", "5.0"],
    ["Piotr", "Wiśniewski", "3.0"],
    ["Katarzyna", "Wójcik", "4.0"]
]

with open("studenci.csv", mode="w", encoding="utf-8", newline="") as plik:
    writer = csv.writer(plik) 
    writer.writerows(dane_studentow)

print("Plik 'studenci.csv' został utworzony!")

# 2. CZYTANIE CSV I ZAPIS DO RAPORTU

suma_ocen = 0
liczba_studentow = 0
najwyzsza_ocena = 0.0
najlepszy_student = ""


with open("studenci.csv", mode="r", encoding="utf-8") as plik:
    reader = csv.reader(plik)
    naglowki = next(reader) 
    
    for wiersz in reader:
        imie = wiersz[0]
        nazwisko = wiersz[1]
        ocena = float(wiersz[2])
        
        suma_ocen += ocena
        liczba_studentow += 1
        if ocena > najwyzsza_ocena:
            najwyzsza_ocena = ocena
            najlepszy_student = f"{imie} {nazwisko}"

srednia = suma_ocen / liczba_studentow if liczba_studentow > 0 else 0

with open("raport.txt", mode="w", encoding="utf-8") as raport:
    raport.write("--- RAPORT OCEN Z PRZEDMIOTU ---\n")
    raport.write(f"Srednia ocen (wszyscy studenci): {srednia:.2f}\n")
    raport.write(f"Najlepszy student: {najlepszy_student} z ocena: {najwyzsza_ocena}\n")

print("Analiza zakończona. Raport zapisano w 'raport.txt'.")
print(f"Średnia ocen wynosi: {srednia:.2f}")
print(f"Najlepszy student to: {najlepszy_student} z oceną {najwyzsza_ocena}")
```

```python
import csv
import os

# FUNKCJE POMOCNICZE / OPCJE MENU

def dodaj_kontakt(kontakty):
    print("\n--- DODAWANIE KONTAKTU ---")
    imie = input("Podaj imię: ").strip()
    nazwisko = input("Podaj nazwisko: ").strip()
    
    while True:
        telefon = input("Podaj numer telefonu (tylko cyfry): ").strip()
        if telefon.isdigit():
            break
        print("Błąd: Numer telefonu musi składać się wyłącznie z cyfr!")
        
    while True:
        email = input("Podaj adres email: ").strip()
        if "@" in email:
            break
        print("Błąd: Nieprawidłowy adres email (brak znaku '@')!")
        
    nowy_kontakt = {
        "imie": imie,
        "nazwisko": nazwisko,
        "telefon": telefon,
        "email": email
    }
    
    kontakty.append(nowy_kontakt)
    print(f"Sukces! Dodano kontakt: {imie} {nazwisko}.")


def wyswietl_kontakty(kontakty):
    print("\n--- LISTA KONTAKTÓW ---")
    if not kontakty:
        print("Książka adresowa jest pusta.")
        return
        
    for i, k in enumerate(kontakty, start=1):
        print(f"{i}. {k['imie']} {k['nazwisko']} | Tel: {k['telefon']} | Email: {k['email']}")


def wyszukaj_kontakt(kontakty):
    print("\n--- WYSZUKIWANIE ---")
    szukana_fraza = input("Podaj imię lub nazwisko (lub fragment): ").strip().lower()
    
    znalezione = []
    for k in kontakty:
        if szukana_fraza in k['imie'].lower() or szukana_fraza in k['nazwisko'].lower():
            znalezione.append(k)
            
    if znalezione:
        print("\nZnalezione kontakty:")
        for k in znalezione:
            print(f"- {k['imie']} {k['nazwisko']} | Tel: {k['telefon']} | Email: {k['email']}")
    else:
        print("Nie znaleziono kontaktów pasujących do tej frazy.")


def usun_kontakt(kontakty):
    print("\n--- USUWANIE KONTAKTU ---")
    wyswietl_kontakty(kontakty)
    
    if not kontakty:
        return
        
    telefon_do_usuniecia = input("\nPodaj numer telefonu kontaktu, który chcesz usunąć: ").strip()
    
    do_usuniecia = None
    for k in kontakty:
        if k['telefon'] == telefon_do_usuniecia:
            do_usuniecia = k
            break
            
    if do_usuniecia:
        kontakty.remove(do_usuniecia) 
        print(f"Sukces! Usunięto: {do_usuniecia['imie']} {do_usuniecia['nazwisko']}.")
    else:
        print("Błąd: Nie ma w bazie kontaktu z takim numerem telefonu.")


def zapisz_kontakty(kontakty, sciezka="kontakty.csv"):
    print("\n--- ZAPIS DO PLIKU ---")
    if not kontakty:
        print("Książka jest pusta. Nie ma czego zapisywać.")
        return
        
    naglowki = ["imie", "nazwisko", "telefon", "email"]
    
    try:
        with open(sciezka, mode="w", newline="", encoding="utf-8") as plik:
            writer = csv.DictWriter(plik, fieldnames=naglowki)
            writer.writeheader()      
            writer.writerows(kontakty) 
        print(f"Zapisano dane do pliku {sciezka}")
    except Exception as e:
        print(f"Błąd podczas zapisywania do pliku: {e}")


def wczytaj_kontakty(sciezka="kontakty.csv"):
    print("\n--- ODCZYT Z PLIKU ---")
    wczytane_kontakty = []
    
    if not os.path.exists(sciezka):
        print("Brak pliku z kontaktami. Zaczynamy z pustą bazą.")
        return wczytane_kontakty
        
    try:
        with open(sciezka, mode="r", encoding="utf-8") as plik:
            reader = csv.DictReader(plik)
            for wiersz in reader:
                wczytane_kontakty.append(dict(wiersz))
        print(f"Wczytano {len(wczytane_kontakty)} kontaktów z pliku {sciezka}.")
    except Exception as e:
        print(f"Błąd podczas odczytu pliku: {e}")
        
    return wczytane_kontakty

# GŁÓWNA PĘTLA PROGRAMU (MENU)

def menu_glowne():
    
    print("Witaj w Menedżerze Kontaktów")
    
    baza_kontaktow = wczytaj_kontakty()

    while True:
        print("\n--- MENU ---")
        print("1. Dodaj kontakt")
        print("2. Wyświetl wszystkie kontakty")
        print("3. Wyszukaj kontakt")
        print("4. Usuń kontakt")
        print("5. Zapisz kontakty do pliku (CSV)")
        print("6. Wczytaj kontakty z pliku (CSV) ręcznie (nadpisze bieżącą listę!)")
        print("7. Wyjdź")
        
        wybor = input("\nWybierz opcję (1-7): ").strip()
        
        if wybor == "1":
            dodaj_kontakt(baza_kontaktow)
        elif wybor == "2":
            wyswietl_kontakty(baza_kontaktow)
        elif wybor == "3":
            wyszukaj_kontakt(baza_kontaktow)
        elif wybor == "4":
            usun_kontakt(baza_kontaktow)
        elif wybor == "5":
            zapisz_kontakty(baza_kontaktow)
        elif wybor == "6":
            baza_kontaktow = wczytaj_kontakty()
        elif wybor == "7":
            pytanie = input("Czy zapisać dane przed wyjściem? (t/n): ").lower()
            if pytanie == "t":
                zapisz_kontakty(baza_kontaktow)
            print("Zamykanie programu. Do widzenia!")
            break
        else:
            print("Nieznana opcja. Wybierz numer od 1 do 7.")

if __name__ == "__main__":
    menu_glowne()
```
